```python
import re

email_str = '''the emails are support@codecademy.com
 or help@codecademy.com'''

emails = re.findall(r"\S+@\S+", email_str)
# \S matches any character except a space, tab or newline

print(emails)
```

```python
from nltk.tokenize import word_tokenize
sent="China is a big country"
token=word_tokenize(sent)
print(token)
```

```python
x=[]
for i in token:
    x.append(i.lower())
print(x)
```

```python
from collections import Counter
tokens=word_tokenize("the cat likes the box.the cat in the box.the cat on the box")
Counter(tokens)
```

```python
Counter(tokens).most_common(3)
```

```python
from nltk.corpus import stopwords
text="the cat likes the box.the cat in the box.the cat on the box"
tokens=[w for w in word_tokenize(text.lower()) if w.isalpha()]
no_stops=[t for t in tokens if t not in stopwords.words("english")]
Counter(no_stops).most_common(2)
```
